```python
cmplist=[]
prev=[]
dis=""
occ=0
for x in dat[1:]:
    
    tem=x.find_all("td")

    if len(tem[0].text.strip())==0 and len(tem[1].text.strip())==0:
        if(tem[2].text.strip()==""):
            pass
        else:
            cmplist.append((dis,occ,tem[2].text.strip()[14:].replace("\n","")))
#             dataset2[dis].append()
    else:   
        prev=[]
        dis=tem[0].text.strip()[14:].replace("\n","")
        occ=int(tem[1].text.strip())
#         dataset1[dis]=occ
        cmplist.append((dis,occ,[tem[2].text.strip()]))    
```

```python
dataset=[]
```

```python

for a in cmplist:
    x,y,z=a
    z="".join(z).replace("\n","")
    if "_" in x:
        x=x.split("_")[-1]
    x=x.replace("\n","")    
    if "_" in z:
        z=z.split("_")[-1]
    dataset.append([x,y,z])
```

```python
data=p.DataFrame(dataset,columns=heading)
```
